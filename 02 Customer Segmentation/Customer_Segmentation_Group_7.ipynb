# Calculate features

```python
df['YEAR_MONTH'] = df['SHOP_DATE'].dt.strftime('%m-%Y')
```

```python
df['SHOP_DATE'].max() # last date
```

```python
l3m_df = df[df['YEAR_MONTH'].isin(['07-2008', '06-2008', '05-2008'])]
l3m_df = l3m_df.groupby(by=['CUST_CODE']).agg(
            l3_TotalSpend=('SPEND', 'sum'),
            l3_AvgSpend=('SPEND', 'mean'),
            l3_MaxSpend=('SPEND', 'max'),
            l3_MinSpend=('SPEND', 'min'),
            l3_ActiveMonth=('YEAR_MONTH', 'nunique'),
            l3_TotalVisits=('BASKET_ID', 'nunique'),
            l3_TotalSKUs=('PROD_CODE', 'nunique'),
        ).reset_index()
l3m_df['l3_TicketSize'] = l3m_df['l3_TotalSpend']/l3m_df['l3_TotalVisits']
l3m_df
```

```python
l6m_df = df[df['YEAR_MONTH'].isin(['07-2008', '06-2008', '05-2008','04-2008','03-2008','02-2008'])]
l6m_df = l6m_df.groupby(by=['CUST_CODE']).agg(
            l6_TotalSpend=('SPEND', 'sum'),
            l6_AvgSpend=('SPEND', 'mean'),
            l6_MaxSpend=('SPEND', 'max'),
            l6_MinSpend=('SPEND', 'min'),
            l6_ActiveMonth=('YEAR_MONTH', 'nunique'),
            l6_TotalVisits=('BASKET_ID', 'nunique'),
            l6_TotalSKUs=('PROD_CODE', 'nunique'),
        ).reset_index()
l6m_df['l6_TicketSize'] = l6m_df['l6_TotalSpend']/l6m_df['l6_TotalVisits']
l6m_df
```

```python
main_df = df[df['CUST_CODE'].notnull()].groupby(by=['CUST_CODE']).agg(
    TotalSpend=('SPEND', 'sum'),
    AvgSpend=('SPEND', 'mean'),
    MaxSpend=('SPEND', 'max'),
    MinSpend=('SPEND', 'min'),
    ActiveMonth=('YEAR_MONTH', 'nunique'),
    TotalVisits=('BASKET_ID', 'nunique'),
    TotalSKUs=('PROD_CODE', 'nunique'),
    FirstDate=('SHOP_DATE', 'min'),
    LastDate=('SHOP_DATE', 'max')
).reset_index()
main_df
```

```python
mtbp_df = df[['CUST_CODE', 'SHOP_DATE']].drop_duplicates().sort_values('SHOP_DATE')
mtbp_df['lag'] = mtbp_df[['CUST_CODE', 'SHOP_DATE']].groupby(['CUST_CODE'])['SHOP_DATE'].shift(1)
mtbp_df['date_diff'] = (mtbp_df['SHOP_DATE'] - mtbp_df['lag'])/np.timedelta64(1,'D')
mtbp_df = mtbp_df.groupby(['CUST_CODE']).agg(mean_time = pd.NamedAgg(column='date_diff', aggfunc='mean')).reset_index().fillna(0)
mtbp_df
```

```python
cst_sgv = main_df.merge(l3m_df,on='CUST_CODE',how='left').merge(l6m_df,on='CUST_CODE',how='left').merge(mtbp_df,on='CUST_CODE',how='left').fillna(0)
cst_sgv
```

```python
##calculate ticket size
cst_sgv['TicketSize'] = cst_sgv['TotalSpend']/cst_sgv['TotalVisits']
```

```python
##find max date in the dataset
max_date = cst_sgv['LastDate'].max()
```

```python
##calculate total days of the relationship
cst_sgv['total_days'] = (cst_sgv['LastDate'] - cst_sgv['FirstDate']).dt.days + 1
```

```python
##calculate recency days
cst_sgv['recency'] = (max_date - cst_sgv['LastDate']).dt.days
```

# Compare model performance

```python
metrics = []
for model in models().index:
    if model not in ['kmeans']:
        continue
    create_model(model)
    metric_result = pull()
    metric_result['model'] = model
    metrics.append(metric_result)
```

```python
cluster_metrics = pd.concat(metrics)
cluster_metrics.set_index("model", inplace=True)
cluster_metrics.sort_values(by='Silhouette', ascending=False, inplace=True)
cluster_metrics.style.highlight_max(subset=['Silhouette', 'Calinski-Harabasz'], color = 'green', axis = 0).highlight_min(subset=['Davies-Bouldin'], color = 'green', axis = 0)
```

```python
kmeans_predictions = predict_model(kmeans, data = cst_sgv)
kmeans_predictions['Cluster'] = kmeans_predictions['Cluster'].str.replace('Cluster ','')
kmeans_predictions
```

```python
feature_cols = kmeans_predictions.drop(['CUST_CODE','Cluster','FirstDate','LastDate'],axis=1).columns
feature_cols = list(feature_cols)
```

```python
kmeans_predictions.head()
```

```python
from matplotlib import pyplot as plt
import graphviz
from sklearn.ensemble import RandomForestClassifier
```

```python
x_tree = kmeans_predictions[feature_cols]
y_tree = kmeans_predictions['Cluster']
clf = tree.DecisionTreeClassifier(random_state=7, max_depth=5)
clf.fit(X=x_tree, y=y_tree)
fig = plt.figure(figsize=(50, 50))
tree.plot_tree(clf, feature_names=x_tree.columns,
                class_names= np.sort(kmeans_predictions['Cluster'].unique()).astype(str),
                filled=True)
plt.show()

dot_data = tree.export_graphviz(clf ,
                                class_names= np.sort(kmeans_predictions['Cluster'].unique()).astype(str),
                                out_file=None,
                                filled=True, rounded=True,
                                feature_names=x_tree.columns)

graph = graphviz.Source(dot_data)
graph.render('plot_tree')
rf = RandomForestClassifier(n_estimators=10)
rf.fit(x_tree, y_tree)
sorted_idx = rf.feature_importances_.argsort()[-10:]
plt.barh(x_tree.columns[sorted_idx], rf.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```
